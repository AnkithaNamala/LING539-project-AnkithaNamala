```python
#importing required libraries
import gensim
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
import re
import emoji
import string
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# downloading particular packages from nltk
nltk.download("punkt")
nltk.download('punkt_tab')
```

```python
# Loading comments from the comments file in the data folder
with open("../data/comments.txt", "r", encoding = "utf-8") as f:
    comments = f.readlines()

print(f"Loaded comments.")
```

```python
# defining clean_comments function that removes emoji characters, removes URLs, hasthags, mentions, punctuation, numbers and extra whitespaces and return the cleaned string
def clean_comments(text):
    unclean_data = re.compile("["u"\U0001F600-\U0001F64F"u"\U0001F300-\U0001F5FF"u"\U00002700-\U000027BF"u"\U000024C2-\U0001F251""]+", flags = re.UNICODE)

    text = unclean_data.sub(r"", text)

    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)

    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)

    text = re.sub(r"\d+", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    text = emoji.replace_emoji(text, replace = "")

    return text
    
# applying the clean_cmments function on the scraped youtube comments
comments_cleaned = [clean_comments(c) for c in comments]
```

```python
#tokenising the comments to pass throuhthe model
def preprocess(text):
    text = text.lower()
    text = re.sub(r"[" + string.punctuation + "]", "", text)
    tokens = word_tokenize(text)

    return tokens

tokenized_cmnts = [preprocess(c) for c in comments_cleaned]
```

```python
# after tokenising, observed that the data still had some emojis amd unwanted characters. hence, deined cleaning_emojis function to remove those
def cleaning_emojis(tokens):
    unwanted_emojis = ["ü•∞", "ü§£", "ü§ó", "ü§£ü§£", "ü§£ü§£ü§£", "ü•∞ü•∞", "ü§©", "ü§î", "üö©", "ü§ù", "ü§¶", "ü´Ç", "ü•∞ü•∞ü•∞", "ü§£ü§£ü§£ü§£ü§£", "ü§£ü§£ü§£ü§£", "ü§¶\u200d", "ü§∑\u200d", "ü§îü§îü§î", "ü§û", "ü§å", "ü•∫", "ü§§ü§§", "ü§°", "\u200d\u200d", "ü•∞ü•∞ü•∞ü•∞ü•∞ü•∞", "ü•≤", "ü§©ü§©", "ü´Çü´Ç", "ü•≤\U0001f979", "üö©üö©", "ü§©ü•∞", "üß°ü§ç", "ü§Ø", "\u200dü©π", "ü§ò", "ü§¶\u200dü§¶\u200dü§¶\u200d", "ü§§", "ü•∞ü•∞ü•∞ü•∞", "üß†"]

    return [t for t in tokens if t not in unwanted_emojis]

cleaned_comments = cleaning_emojis(tokenized_cmnts)
```

```python
# saving this cleaned comments list as cleaned_comments text file in the data folder
with open("../data/cleaned_comments.txt", "w", encoding = "utf-8") as f:
    for comment in cleaned_comments:
        l = " ".join(comment)
        f.write(l + "\n")

print("Comments saved to cleaned_comments.txt file")
```

```python
# training the model
skipgram_model = Word2Vec(sentences = cleaned_comments,
                          vector_size = 100,
                          window = 5,
                          min_count = 2,
                          sg = 1,
                          workers = 4,
                          epochs = 10)
```

```python
# to view the model use this command: print(skipgram_model.wv.index_to_key)
```

```python
# saving the model for future use

skipgram_model.save("telugu_english_skipgram.model")
```

```python
# defining visualise_word_embeddings function to visualise top 5 word embeddings and print word and similarity scores.
model = Word2Vec.load("telugu_english_skipgram.model")

def visualise_word_embeddings(word):

    similar_words = model.wv.most_similar(word, topn = 5)

    print(f"Words similar to {word}:")

    for w, s in similar_words:
        print(f"{w}: {s:.4f}")

    words = [w for w,s in similar_words]
    vectors = [model.wv[w] for w in words]

    words.insert(0, word)
    vectors.insert(0, model.wv[word])

    pca = PCA(n_components = 2)
    res = pca.fit_transform(vectors)

    plt.figure(figsize = (5, 5))
    plt.scatter(res[:, 0], res[:, 1])

    for i, w in enumerate(words):
        plt.annotate(w, xy = (res[i, 0], res[i, 1]))

    plt.title(f"2D PCA Word Embeddings {word}")
    plt.show()
```

```python
#trying to see how data is captured for some example words
example_words = ["skit","cinema", "enti", "repu", "super"]

for word in example_words:
    visualise_word_embeddings(word)
```
